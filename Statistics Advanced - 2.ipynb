```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of random samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Given values
mu_0 = 50   # population mean
sigma = 0.5 # population standard deviation
alpha = 0.05

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# P-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from 50.")
else:
    print("Fail to reject the null hypothesis: No significant difference from 50.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, sd=5, n=100

# Step 2: Calculate statistics
mean = np.mean(data)
std_err = stats.sem(data)  # standard error of mean
confidence_level = 0.95
confidence_interval = stats.t.interval(confidence_level, len(data)-1, loc=mean, scale=std_err)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")

# Step 3: Plot data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f"Lower CI = {confidence_interval[0]:.2f}")
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f"Upper CI = {confidence_interval[1]:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev

# Sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)

# Calculate Z-scores
z_scores, mean, std_dev = calculate_z_scores(data)

print(f"Original Mean: {mean:.2f}")
print(f"Original Std Dev: {std_dev:.2f}")
print(f"First 5 Z-scores: {z_scores[:5]}")

# Plot histogram of Z-scores
plt.figure(figsize=(8,5))
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.axvline(1, color='green', linestyle='--', label="Z=+1 (1 Std Dev)")
plt.axvline(-1, color='green', linestyle='--', label="Z=-1 (1 Std Dev)")
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```
